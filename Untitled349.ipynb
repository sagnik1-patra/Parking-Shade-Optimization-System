```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# PATH
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Parking Shade Optimization System\archive"
os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD DATA
# ============================================================

def melt_df(df, value_name):
    return df.melt(id_vars=["datetime"], var_name="city", value_name=value_name)

humidity = melt_df(pd.read_csv(os.path.join(BASE_DIR,"humidity.csv")), "humidity")
pressure = melt_df(pd.read_csv(os.path.join(BASE_DIR,"pressure.csv")), "pressure")
temperature = melt_df(pd.read_csv(os.path.join(BASE_DIR,"temperature.csv")), "temperature")
wind_speed = melt_df(pd.read_csv(os.path.join(BASE_DIR,"wind_speed.csv")), "wind_speed")
wind_dir = melt_df(pd.read_csv(os.path.join(BASE_DIR,"wind_direction.csv")), "wind_direction")

df = humidity.merge(pressure,on=["datetime","city"]) \
             .merge(temperature,on=["datetime","city"]) \
             .merge(wind_speed,on=["datetime","city"]) \
             .merge(wind_dir,on=["datetime","city"])

df.dropna(inplace=True)

# ============================================================
# FEATURE ENGINEERING
# ============================================================

df["shade_need_index"] = (
    df["temperature"]*0.5 +
    df["humidity"]*0.2 -
    df["wind_speed"]*0.3
)

features = ["humidity","pressure","temperature","wind_speed","wind_direction"]
target = "shade_need_index"

X = df[features].values
y = df[target].values

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ============================================================
# CNN MODEL
# ============================================================

model = Sequential([
    Conv1D(32,2,activation="relu",input_shape=(X_scaled.shape[1],1)),
    Conv1D(64,2,activation="relu"),
    Flatten(),
    Dense(64,activation="relu"),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

history = model.fit(
    X_train, y_train,
    epochs=25,
    batch_size=64,
    validation_split=0.1,
    callbacks=[EarlyStopping(patience=5)],
    verbose=1
)

# ============================================================
# EVALUATION
# ============================================================

preds = model.predict(X_test).flatten()
mse = mean_squared_error(y_test, preds)
r2 = r2_score(y_test, preds)

# ============================================================
# 1 ACCURACY / LOSS GRAPH
# ============================================================

plt.figure()
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.title("CNN Training Loss")
plt.savefig(os.path.join(BASE_DIR,"accuracy_loss.png"))
plt.show()

# ============================================================
# 2 COMPARISON GRAPH
# ============================================================

plt.figure()
plt.scatter(y_test, preds, alpha=0.5)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Shade Index")
plt.savefig(os.path.join(BASE_DIR,"comparison_graph.png"))
plt.show()

# ============================================================
# 3 HEATMAP
# ============================================================

plt.figure(figsize=(8,6))
sns.heatmap(df[features+[target]].corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR,"heatmap.png"))
plt.show()

# ============================================================
# 4 RESULT CSV
# ============================================================

results_df = pd.DataFrame({
    "actual": y_test,
    "predicted": preds,
    "error": y_test - preds
})

results_df.to_csv(
    os.path.join(BASE_DIR,"parking_shade_results.csv"),
    index=False
)

# ============================================================
# 5 RESULT ERROR GRAPH
# ============================================================

plt.figure()
plt.hist(results_df["error"], bins=50)
plt.title("Prediction Error Distribution")
plt.savefig(os.path.join(BASE_DIR,"error_distribution.png"))
plt.show()

# ============================================================
# 6 PREDICTION JSON
# ============================================================

prediction_json = {
    "mse": float(mse),
    "r2_score": float(r2),
    "sample_predictions": preds[:20].tolist()
}

with open(os.path.join(BASE_DIR,"parking_shade_predictions.json"),"w") as f:
    json.dump(prediction_json,f,indent=4)

# ============================================================
# 7 PREDICTION GRAPH
# ============================================================

plt.figure()
plt.plot(preds[:100], label="Predicted")
plt.plot(y_test[:100], label="Actual")
plt.legend()
plt.title("Prediction Trend (First 100)")
plt.savefig(os.path.join(BASE_DIR,"prediction_graph.png"))
plt.show()

# ============================================================
# SAVE MODEL & OBJECTS
# ============================================================

model.save(os.path.join(BASE_DIR,"parking_shade_cnn_model.h5"))

with open(os.path.join(BASE_DIR,"parking_shade_scaler.pkl"),"wb") as f:
    pickle.dump(scaler,f)

with open(os.path.join(BASE_DIR,"parking_shade_config.yaml"),"w") as f:
    yaml.dump({
        "model":"CNN + PSO",
        "features":features,
        "mse":float(mse),
        "r2_score":float(r2)
    },f)

print(" ALL RESULTS GENERATED & SAVED SUCCESSFULLY")
```
