```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten
from tensorflow.keras.callbacks import EarlyStopping

# ============================================================
# PATHS
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Parking Shade Optimization System"
DATA_DIR = os.path.join(BASE_DIR, "archive")
os.makedirs(BASE_DIR, exist_ok=True)

# ============================================================
# LOAD & MERGE DATA
# ============================================================

def melt_df(df, name):
    return df.melt(id_vars=["datetime"], var_name="city", value_name=name)

humidity = melt_df(pd.read_csv(os.path.join(DATA_DIR,"humidity.csv")), "humidity")
pressure = melt_df(pd.read_csv(os.path.join(DATA_DIR,"pressure.csv")), "pressure")
temperature = melt_df(pd.read_csv(os.path.join(DATA_DIR,"temperature.csv")), "temperature")
wind_speed = melt_df(pd.read_csv(os.path.join(DATA_DIR,"wind_speed.csv")), "wind_speed")
wind_dir = melt_df(pd.read_csv(os.path.join(DATA_DIR,"wind_direction.csv")), "wind_direction")

df = humidity.merge(pressure,on=["datetime","city"]) \
             .merge(temperature,on=["datetime","city"]) \
             .merge(wind_speed,on=["datetime","city"]) \
             .merge(wind_dir,on=["datetime","city"])

df.dropna(inplace=True)

# ============================================================
# FEATURE ENGINEERING
# ============================================================

df["shade_need_index"] = (
    df["temperature"] * 0.5 +
    df["humidity"] * 0.2 -
    df["wind_speed"] * 0.3
)

features = ["humidity","pressure","temperature","wind_speed","wind_direction"]
target = "shade_need_index"

X = df[features].values
y = df[target].values

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# ============================================================
# CNN MODEL
# ============================================================

model = Sequential([
    Conv1D(32,2,activation="relu",input_shape=(X_scaled.shape[1],1)),
    Conv1D(64,2,activation="relu"),
    Flatten(),
    Dense(64,activation="relu"),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

history = model.fit(
    X_train, y_train,
    epochs=25,
    batch_size=64,
    validation_split=0.1,
    callbacks=[EarlyStopping(patience=5)],
    verbose=1
)

# ============================================================
# EVALUATION
# ============================================================

preds = model.predict(X_test).flatten()
mse = mean_squared_error(y_test, preds)
r2 = r2_score(y_test, preds)

# ============================================================
# FITNESS FUNCTION
# ============================================================

def fitness(sol):
    return np.sum(sol * np.mean(X, axis=0))

# ============================================================
# AIS (FIXED)
# ============================================================

def AIS(pop=20, dim=len(features), iterations=30):
    memory = np.random.rand(pop, dim)
    for _ in range(iterations):
        noise = np.random.normal(0, 0.1, memory.shape)
        clones = memory + noise
        scores = np.array([fitness(c) for c in clones])
        best_idx = np.argsort(scores)[:pop]
        memory = clones[best_idx]
    return memory[0]

# ============================================================
# PSO
# ============================================================

def PSO(initial, particles=30, iterations=50):
    dim = len(initial)
    swarm = np.random.rand(particles, dim)
    velocity = np.zeros_like(swarm)

    pbest = swarm.copy()
    pbest_score = np.array([fitness(p) for p in pbest])

    gbest = initial.copy()
    gbest_score = fitness(gbest)

    for _ in range(iterations):
        for i in range(particles):
            r1, r2 = random.random(), random.random()
            velocity[i] = (
                0.5 * velocity[i]
                + r1 * (pbest[i] - swarm[i])
                + r2 * (gbest - swarm[i])
            )
            swarm[i] += velocity[i]

            score = fitness(swarm[i])
            if score < pbest_score[i]:
                pbest[i] = swarm[i]
                pbest_score[i] = score

        if np.min(pbest_score) < gbest_score:
            gbest = pbest[np.argmin(pbest_score)]
            gbest_score = np.min(pbest_score)

    return gbest

# ============================================================
# HYBRID AIS  PSO
# ============================================================

ais_best = AIS()
pis_best = PSO(ais_best)

# ============================================================
# VISUALIZATIONS
# ============================================================

# Accuracy / Loss
plt.figure()
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.legend()
plt.title("Hybrid AIS + PSO Training Loss")
plt.savefig(os.path.join(BASE_DIR,"pis_accuracy_loss.png"))
plt.show()

# Actual vs Predicted
plt.figure()
plt.scatter(y_test, preds, alpha=0.5)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Hybrid AIS + PSO: Actual vs Predicted")
plt.savefig(os.path.join(BASE_DIR,"pis_comparison_graph.png"))
plt.show()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[features+[target]].corr(), annot=True, cmap="coolwarm")
plt.title("Hybrid AIS + PSO Correlation Heatmap")
plt.savefig(os.path.join(BASE_DIR,"pis_heatmap.png"))
plt.show()

# Error Distribution
error = y_test - preds
plt.figure()
plt.hist(error, bins=50)
plt.title("Hybrid AIS + PSO Error Distribution")
plt.savefig(os.path.join(BASE_DIR,"pis_error_distribution.png"))
plt.show()

# Prediction Trend
plt.figure()
plt.plot(preds[:100], label="Predicted")
plt.plot(y_test[:100], label="Actual")
plt.legend()
plt.title("Hybrid AIS + PSO Prediction Trend")
plt.savefig(os.path.join(BASE_DIR,"pis_prediction_graph.png"))
plt.show()

# ============================================================
# SAVE OUTPUTS
# ============================================================

# CSV
pd.DataFrame({
    "actual": y_test,
    "predicted": preds,
    "error": error
}).to_csv(
    os.path.join(BASE_DIR,"pis_parking_shade_results.csv"),
    index=False
)

# JSON
with open(os.path.join(BASE_DIR,"pis_parking_shade_predictions.json"),"w") as f:
    json.dump({
        "mse": float(mse),
        "r2_score": float(r2),
        "ais_best": ais_best.tolist(),
        "pis_best": pis_best.tolist(),
        "sample_predictions": preds[:20].tolist()
    }, f, indent=4)

# MODEL
model.save(os.path.join(BASE_DIR,"pis_parking_shade_model.h5"))

# OBJECTS
with open(os.path.join(BASE_DIR,"pis_parking_shade_objects.pkl"),"wb") as f:
    pickle.dump({
        "scaler": scaler,
        "ais_best": ais_best,
        "pis_best": pis_best
    }, f)

# CONFIG
with open(os.path.join(BASE_DIR,"pis_parking_shade_config.yaml"),"w") as f:
    yaml.dump({
        "model": "CNN + Hybrid AIS + PSO",
        "features": features,
        "mse": float(mse),
        "r2_score": float(r2)
    }, f)

print(" HYBRID AIS + PSO (pis_) PIPELINE COMPLETED SUCCESSFULLY")
```
